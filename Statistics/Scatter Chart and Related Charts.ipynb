```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.describe()
```

```python
g = sns.PairGrid(data=iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(data=iris)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

```python
sns.scatterplot(x='sepal_width',y='sepal_length',data=iris)
```

```python
g = sns.jointplot(x='sepal_width',y='sepal_length',data=iris)
g.plot(sns.regplot,sns.histplot)
```
