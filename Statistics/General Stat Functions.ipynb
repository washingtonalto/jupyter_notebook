```python
# General Stat Functions
#
# Created by: Washington Alto
# Date: Mar 28, 2022

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline
```

```python
def mystatfunc(x,functype, **kwargs):
    if functype == 'normal':
        y = stats.norm.pdf(x,kwargs['loc'],kwargs['scale'])
    elif functype == 'T':
        y = stats.t.pdf(x,kwargs['df'],kwargs['loc'],kwargs['scale'])
    elif functype == 'f':
        y = stats.f.pdf(x,kwargs['dfn'],kwargs['dfd'],kwargs['loc'],kwargs['scale'])
    else:
        pass
    return y
```

```python
# Parameters for the curve
x_min = -3.0
x_max = 3.0


mean = 0 
std = 1.0
df = 2
dfn = 3
dfd = 3

x = np.linspace(x_min, x_max, 100)


fig,axes = plt.subplots(figsize=(12,8))
plt.title('Probability Distribution Function')
plt.xlabel('x')
plt.plot(x,mystatfunc(x,'normal',loc=mean,scale=std),color='blue',linewidth=1,label='Normal - μ:{mean:.2f},σ: {std:.2f}'.format(mean=mean,std=std))
plt.plot(x,mystatfunc(x,'T',df=df,loc=mean,scale=std),color='green',linewidth=1,label='T - df:{df}, μ:{mean:.2f},σ: {std:.2f}'.format(mean=mean,std=std,df=df))
plt.plot(x,mystatfunc(x,'f',dfd=dfd,dfn=dfn,loc=mean,scale=std),color='red',linewidth=1,label='f - dfd:{dfd}, dfn:{dfn}, μ:{mean:.2f},σ: {std:.2f}'.format(mean=mean,std=std,dfd=dfd,dfn=dfn))
plt.ylabel('P(x)')
_ = plt.legend()
```
