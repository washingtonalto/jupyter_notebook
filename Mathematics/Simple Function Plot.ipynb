```python
# Simple Function Plot
#
# Created by: Washington Alto
# Date: Sep 15, 2022

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
def plot_function(formulatext,legendformulatext,scale=5,functionlinecolor='blue'):
# Parameters: formulatext - string representing the right-hand-side (RHS) formula used in the function. Note that 
#                           function calls should use Numpy function calls such as np.sqrt,np.sin and so forth.
#                           See https://numpy.org/doc/stable/reference/routines.math.html 
#                           Also, functions calls can also use Scipy Stats e.g. scipy.stats functions. 
#             legendformulatext - string representing the legend for the function to be used. If latex is to be used
#                           to represent the function use raw string with '$'. For example,
#                           r'$f(x) = \frac{-2x}{\sqrt{x^2 + 3}}$'
#             scale - represent the absolute number for minimum and maximum x and y scale. By default, this is 5
#             functionlinecolor - color for the line represented by the function. By default, color is 'blue' 

    # function to plot on the line chart
    def myfunc(x):
        # CAUTION: Use eval function with care! Eval is a powerful yet if used wrong, it is a dangerous function 
        # as it can be misused to hack. Used only for local machine jupyter notebook operation
        y = eval(formulatext)
        return y
    
    # Set the minimum and maximum values of both x and y based on the scale 
    xmin = -scale
    xmax = scale
    ymin = -scale
    ymax = scale

    # Set the X and Y vector values used for plotting the function f(x)
    x = np.linspace(xmin,xmax,scale*100)
    y = myfunc(x)
    
    # plot routines
    plt.figure(figsize=(8,6),dpi=150)
    plt.plot(x,y,label=legendformulatext,color=functionlinecolor)
    plt.hlines(y=0,xmin=xmin,xmax=xmax,colors='black')
    plt.vlines(x=0,ymin=ymin,ymax=ymax,colors='black')
    plt.xlim(xmin,xmax)
    plt.ylim(ymin,ymax)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    if scale < 12:
        plt.xticks(np.arange(xmin,xmax,1))
        plt.yticks(np.arange(ymin,ymax,1))
    plt.grid(linestyle = '--', linewidth = 0.40)
    plt.legend(loc='lower right');
```

```python
plot_function("-2*x/np.sqrt(x**2 + 3)",r'$f(x) = \frac{-2x}{\sqrt{x^2 + 3}}$')
```

```python
plot_function("x**2/(x**2 + 4)",r'$f(x) = \frac{x^2}{x^2 + 4}$')
```

```python
plot_function("4*(x**2)/(x**2 + 3)",r'$f(x) = \frac{4x^2}{x^2 + 3}$')
```

```python
plot_function("x**3 - 3*(x**2) + 3",r'$f(x) = x^3 - 3x^2 + 3$')
```

```python
plot_function("np.sin(x)",'f(x) = sin(x)',functionlinecolor='green')
```

```python
plot_function("stats.norm.pdf(x,0,1)",'f(x) = norm(x,0,1)',scale=4)
```
