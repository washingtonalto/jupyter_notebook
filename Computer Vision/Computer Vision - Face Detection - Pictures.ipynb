```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_picture(filename):
    img = cv2.imread(filename)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_fullpicture(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
def display_sidebysidepicture(img1,img2):
    fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10,8))
    axes[0].imshow(img1)
    axes[1].imshow(img2)
```

```python
def detect_object(img,cascade,scaleFactor=1.2,label='Face detected',minNeighbors=5,rectColor=(0,255,0),rectLinethickness=5,fontScale=2):
    
  
    face_img = img.copy()
    face_rects = cascade.detectMultiScale(face_img,scaleFactor,minNeighbors) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h),rectColor, rectLinethickness) 
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(face_img,text=label,org=(x,y - 10), fontFace=font,fontScale=fontScale,color=rectColor,thickness=2,lineType=cv2.LINE_AA)

        
    return face_img
    
```

```python
catface_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalcatface.xml')
humanface_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
```

**Note:** [Denis Muwege](https://www.nobelprize.org/prizes/peace/2018/mukwege/facts/) is a world-renowned gynaecologist, human rights activist and Nobel Peace Prize laureate from east Congo. He has become the worldâ€™s leading specialist in the treatment of wartime sexual violence and a global campaigner against the use of rape as a weapon of war.

```python
humanface = load_picture('DATA/Denis_Mukwege.jpg')
humanface_detect = detect_object(humanface,humanface_cascade)
display_sidebysidepicture(humanface,humanface_detect)
```

**Note:** [Nadia Murad Basee Taha](https://en.wikipedia.org/wiki/Nadia_Murad) is an Iraqi Yazidi human rights activist who lives in Germany.

```python
humanface = load_picture('DATA/Nadia_Murad.jpg')
humanface_detect = detect_object(humanface,humanface_cascade,fontScale=1)
display_sidebysidepicture(humanface,humanface_detect)
```
